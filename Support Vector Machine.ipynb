```python
# Support Vector Machine 
import pandas as pd
from sklearn.datasets import load_iris
iris= load_iris()

```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
# Loading dataset into df
# with columns = iris.feature_names, we get column names
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
df
```

```python
iris.target
```

```python
df['target'] = iris.target
```

```python
df
```

```python
# There are 3 classes: c1,c2,c3
iris.target_names
```

```python
df[df.target==1]
```

```python
# Replacing numbers with flower name [0,1,2]== ['setosa', 'versicolor', 'virginica]
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
# To visualize data

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df2
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'red', marker= '+')

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color= 'blue', marker = '+')

plt.show()
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'red', marker= '+')

plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color= 'blue', marker = '+')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower_name'], axis='columns')
```

```python
y = df['target']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state = 42)
```

```python
x_train
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# creating model 
# Shift+ tab to see list of kernels
from sklearn.svm import SVC
reg = SVC(kernel ='linear')
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
reg.predict(x_test)
```
